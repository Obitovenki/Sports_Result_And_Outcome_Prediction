<h1><center>Football Exploratory Data Analysis</center></h1>

This Notebook will be taking a look at Football Data, the data will specifically be looking at the English Premier League and the data will be data on the players in which various stats are taken the season this data will be exploring will be the 2020/21 season. The notebook will make use of the Pandas, numpy and many other Python libraries to analyse the data on this season.

```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load the data
epl_df = pd.read_csv("Dataset/EPL_20_21.csv")

#Print the first five rows
epl_df.head()
```

```python
#Display information on the epl_df
epl_df.info()
```

```python
#Display descriptive data for epl_df
epl_df.describe()
```

```python
#Check how many null values are in the columns
epl_df.isnull().sum()
```

```python
#Another way to check for null values in the dataframe
epl_df.isna().sum()
```

```python
#display the columns in the dataframe
epl_df.columns
```

```python
#Create new columns
epl_df['MinsPerMatch'] = (epl_df['Mins'] / epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch'] = (epl_df['Goals'] / epl_df['Matches']).astype(float)
epl_df['AssistsPerMatch'] = (epl_df['Assists'] / epl_df['Matches']).astype(float)

epl_df.head()
```

```python
#Total amount of Goals scored in the 2020/21 season
Total_Goals = epl_df['Goals'].sum()
Total_Goals
```

```python
#Total amount of Assists made in the 2020/21 season
Total_Assists = epl_df['Assists'].sum()
Total_Assists
```

```python
#Total amount of penalty goals scored in the 2020/21 season
Total_PenaltyGoals = epl_df['Penalty_Goals'].sum()
Total_PenaltyGoals
```

```python
#Total amount of Penaltys attempted in the 2020/21 season
Total_PenaltyAttempts = epl_df['Penalty_Attempted'].sum()
Total_PenaltyAttempts
```

```python
#Pie Chart for penalties Missed vs Scored
plt.figure(figsize=(14,8))
py_not_scored = epl_df['Penalty_Attempted'].sum() - Total_PenaltyGoals
py_data = [py_not_scored, Total_PenaltyGoals]
labels = ['Penalties missed', 'Penalties Scored']
color = sns.color_palette('Set2')
plt.pie(py_data, labels = labels, colors=color, autopct= '%.0f%%')
plt.show()
```

```python
#Unique Positions
epl_df['Position'].unique()
```

```python
#Unique Teams in the dataframe
epl_df['Club'].unique()
```

```python
#Unique nations in the premier league 
epl_df['Nationality'].unique()
```

```python
#Total FW Players
epl_df[epl_df['Position'] == 'FW']
```

```python
#Players from different nations
np.size((epl_df['Nationality'].unique()))
```

```python
#Shows most popular nations in the english premier league 2020/21
pop_nation = epl_df.groupby('Nationality').size().sort_values(ascending=False)

pop_nation.head(10).plot(kind='bar', figsize=(12,6), color=sns.color_palette("pastel"))
plt.title("Most Popular nationalities in the EPL 2022/23 season", fontsize=15)
```

```python
#Most popular positions in the english premier league season 2020/21
pop_position = epl_df.groupby('Position').size().sort_values(ascending=False)

pop_position.head(10).plot(kind='bar', figsize=(12, 6), color=sns.color_palette('Paired'))
plt.title("Most Popular positions in the EPL 2022/23 season", fontsize=15)
```

```python
#Clubs with maximum players in their squad
epl_df['Club'].value_counts().nlargest(5).plot(kind='bar', color=sns.color_palette('crest'))
plt.title("Clubs with the most of amount of Players in their squad", fontsize=20)
```

```python
#Clubs with the least amount of players in their squad
epl_df['Club'].value_counts().nsmallest(5).plot(kind='bar', color=sns.color_palette('magma'))
plt.title("Clubs with the least amount of Players in their squad", fontsize=20)
```

```python
#Players based on the age group
under20 = epl_df.loc[epl_df['Age'] <= 20]
age20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
age25_30 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
above30  = epl_df[epl_df['Age'] > 30]
```

```python
x = np.array([under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), above30['Name'].count()])
myLabels = ["<=20", ">20 & <= 25", "> 25 & <=30", "> 30"]
plt.title('Total Players with Age Groups', fontsize=20)
plt.pie(x, labels=myLabels, autopct="%.1f%%")
plt.show()
```

```python
#Total under 20 players in each club
players_under_20 = epl_df[epl_df['Age'] < 20]
players_under_20['Club'].value_counts().plot(kind ='bar', color=sns.color_palette("cubehelix"), figsize=(18, 8))
plt.title("Clubs with the most players under 20 in the 2022/23 season", fontsize=20)
```

```python
#Players under 20 that played for Arsenal
players_under_20[players_under_20['Club'] =='Arsenal']
```

```python
#Average age of Players at each club
plt.figure(figsize=(12, 6))
sns.boxplot(x ='Club', y='Age', data =epl_df)
plt.xticks(rotation = 90)
plt.title("The average age of Players at each Club in the 2022/23 season", fontsize=15)
```

```python
num_player = epl_df.groupby('Club').size()
avg_player_age = (epl_df.groupby('Club')['Age'].sum()) / num_player
avg_player_age.sort_values(ascending=False)
```

```python
Assists_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Club', y='Assists', data=Assists_by_clubs.sort_values(by='Assists'), palette='Set2')
ax.set_xlabel("Club", fontsize=30)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']=(20, 8)
plt.title("Total assists by club during the 2022/23 season", fontsize=25)
```

```python
#Top 10 Assists
top_10_assists = epl_df[['Name', 'Club', 'Matches', 'Assists']].nlargest(n=10, columns = 'Assists')
top_10_assists
```

```python
Goals_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Club', y='Goals', data=Goals_by_clubs.sort_values(by="Goals"), palette='rocket')
ax.set_xlabel("Club", fontsize=30)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']=(20, 8)
plt.title("Total Goals by clubs during the 2022/23 season", fontsize=25)
```

```python
#Most goals by Players
#Top 10 Assists
top_10_goals = epl_df[['Name', 'Club', 'Matches', 'Goals']].nlargest(n=10, columns = 'Goals')
top_10_goals
```

```python
#Top 10 Goals Per match
top_10_goals_per_match = epl_df[['Name', 'Club', 'Matches', 'GoalsPerMatch']].nlargest(n=10, columns='GoalsPerMatch')
top_10_goals_per_match
```

```python
#Top 10 assists per match
top_10_assists_per_match = epl_df[['Name', 'Club', 'Matches', 'AssistsPerMatch']].nlargest(n=10, columns='AssistsPerMatch')
top_10_assists_per_match
```

```python
#Top 10 goalscorers under the age of 20
top_10_U20_goals = epl_df[['Name', 'Age', 'Club', 'Matches', 'Goals']].loc[epl_df['Age'] < 20].nlargest(n=10, columns='Goals')
top_10_U20_goals
```

```python
#Top 10 assisters under the age of 20
top_10_U20_assists = epl_df[['Name', 'Age', 'Club', 'Matches', 'Assists']].loc[epl_df['Age'] < 20].nlargest(n=10, columns='Assists')
top_10_U20_assists
```

```python
plt.figure(figsize=(14, 7))
assists = epl_df['Assists'].sum()
goals_no_assist_data = [Total_Goals - assists, assists]
labels = ['Goals without assists', 'Goals with assists']
color = sns.color_palette('Set1')
plt.pie(goals_no_assist_data, labels=labels, colors=color, autopct='%.0f%%')
plt.show()
```

```python
#Top 10 players with most yellow cards
epl_yellow = epl_df.sort_values(by='Yellow_Cards', ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title("Players with the most Yellow Cards in 2022/23 season", fontsize=20)
c = sns.barplot(x=epl_yellow['Name'], y=epl_yellow['Yellow_Cards'], label='Players', color='yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(), rotation=45)
```

```python
#Top 10 players with most yellow cards
epl_red = epl_df.sort_values(by='Red_Cards', ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title("Players with the most Red Cards in 2022/23 season", fontsize=20)
r = sns.barplot(x=epl_red['Name'], y=epl_red['Red_Cards'], label='Players', color='red')
plt.ylabel('Number of Red Cards')
r.set_xticklabels(r.get_xticklabels(), rotation=45)
```
